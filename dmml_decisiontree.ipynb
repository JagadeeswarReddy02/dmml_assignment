## ***Reading Data***

```python
df = pd.read_csv('Wholesale_customers_data.csv')
df.head()
```

## ***Exploring Data***

```python
X = df.drop('Region', axis='columns')
y = df['Region']
print(X.shape)
print(y.shape)
```

```python
X = pd.get_dummies(X)
print(X.columns)
feature_names = X.columns
X.head()
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, stratify=y, random_state=2)
```

```python
X=df.drop(['Region'],axis='columns')
y=df['Region']
X.head()
```

```python
for d in range(2,12):
    tree = DecisionTreeClassifier(max_depth=d)
    scores = cross_val_score(tree, X_train, y_train, cv=5)
    print("Depth: ", d, "Accuracy:", scores.mean())
```

```python
tree = DecisionTreeClassifier(max_depth=4)
tree = tree.fit(X_train,y_train)

#Predict the response for test dataset
y_hat = tree.predict(X_test)
print("Test Accuracy: ", tree.score(X_test, y_test))
```

## ***Confusion Matrix***

```python
import numpy as np
cm = confusion_matrix(y_test, y_hat)
print("CM", cm)

# Loop through each class
for i in range(cm.shape[0]):
    tn, fp, fn, tp = cm[i, i], np.sum(cm[i, :]) - cm[i, i], np.sum(cm[:, i]) - cm[i, i], cm[i, i]
    print(f"Class {i}: TN = {tn}, FP = {fp}, FN = {fn}, TP = {tp}")
```

```python
from IPython.display import Image
class_names = [str(cls) for cls in tree.classes_]  # Convert class labels to strings
feature_names = ['feature1', 'feature2', 'feature3', 'feature4', 'feature5', 'feature6', 'feature7'] 

dot_data = StringIO()
export_graphviz(tree, out_file=dot_data,
                filled=True, rounded=True,
                special_characters=True, class_names=class_names, feature_names=feature_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.set_dpi(300)
graph.set_size('"50,50!"')
graph.write_png(r"C:\Users\kurap\Downloads\plots\tree_plot.png")
Image(graph.create_png())
```
